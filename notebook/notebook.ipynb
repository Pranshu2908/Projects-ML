# SCALING FEATURES

```python
data
```

```python
# Create preprocessor ColumnTransformer to do OneHotEncoder for CATEGORIC_FEATURES and StandardScaler() for NUMERIC_FEATURES

num_pipeline = Pipeline([
    ('num_features', StandardScaler())
])

cat_pipeline = Pipeline([
    ('cat_features', OneHotEncoder())
])

# Create the ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num_trans', num_pipeline, numeric_data),
        ('cat_trans', cat_pipeline, categories)
])


data_transformed = preprocessor.fit_transform(data)

# Converting the transformed data back to a dataframe for easier visualization
# The transformed data will have new column names, especially for the one hot encoded categories
encoded_feature_names = preprocessor.named_transformers_['cat_trans'].get_feature_names_out(categories)
new_column_names = list(numeric_data) + list(encoded_feature_names)

data_transformed = pd.DataFrame(data_transformed, columns=new_column_names)

data_transformed.head()
```

# MODELLING

```python


def results(y_true,y_pred):
    f1_score_classwise = f1_score(y_true,y_pred,average=None)
    
    res = {
        'Accuracy': accuracy_score(y_true,y_pred),
        'Recall': recall_score(y_true,y_pred,average='macro'),
        'Precision':precision_score(y_true,y_pred,average='macro'),
        'F1 Score':f1_score(y_true,y_pred,average='macro'),
        'F1 Score By class':f1_score_classwise
    }
    return res
```

#### here i have used average as macro because we have an imbalanced distribution where majority is of no failure class. so in macro it would calculate mean of all the classes so it would give equal importance to each class and not the majority one. 

### train test split

```python

data_model = data.copy()
X = data_model[numeric_data + categories]
y = data_model['failure_type']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=101,test_size=0.2)
```

### model with class weights

```python
model_pipe = Pipeline([
    ('preprocessor',preprocessor),
    ('model',GradientBoostingClassifier(random_state=100))
])

weights =compute_sample_weight(class_weight='balanced', y=y_train)
model_pipe.fit(X_train, y_train, model__sample_weight=weights)

y_pred = model_pipe.predict(X_test)

result = results(y_test,y_pred)
result
```

### model with SMOTE-NC(synthetic minority over-sampling techniques for nomianl and continuous)

```python
categorical_features_indices = [i for i, feature in enumerate(numeric_data + categories) 
                                if feature in categories]

smotenc = SMOTENC(categorical_features=categorical_features_indices, random_state=42)
model_pipe_smotenc = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smotenc', 'passthrough'),
    ('model', GradientBoostingClassifier(random_state=100))
])
model_pipe_smotenc.fit(X_train, y_train)

y_pred_smotenc = model_pipe_smotenc.predict(X_test)
result_smotenc = results(y_test,y_pred_smotenc)
result_smotenc
```

### KNN

```python
smotenc = SMOTENC(categorical_features=categorical_features_indices, random_state=42)
model_pipe_smotenc_k = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smotenc', 'passthrough'),
    ('model', KNeighborsClassifier())
])
model_pipe_smotenc_k.fit(X_train, y_train)
y_pred_smotenc_k = model_pipe_smotenc_k.predict(X_test)
result_smotenc_k = results(y_test,y_pred_smotenc_k)
result_smotenc_k
```

### gridsearch for finding best model

```python
params = {
    'model': [
        LogisticRegressionCV(max_iter=500, random_state=100),
        RandomForestClassifier(random_state=100),
        GradientBoostingClassifier(random_state=100),
        KNeighborsClassifier()
    ]
}

grid = GridSearchCV(model_pipe_smotenc, params, cv=3, scoring='f1_macro', n_jobs=-1, verbose=1)
grid.fit(X_train, y_train)

result_grid = pd.DataFrame(grid.cv_results_)
best_model_index = result_grid['mean_test_score'].idxmax()
best_model_params = result_grid.loc[best_model_index, 'params']

best_model_params
```

### lets perform hyperparamter tunning

```python
#help(GradientBoostingClassifier)
```

```python
param = {
    'model__n_estimators':[10,50,150,200,300,450,500],
    'model__max_depth':[None,1,5,10,15]
}

fine_tune_grid = GridSearchCV(model_pipe_smotenc,param, cv=3, scoring='f1_macro', n_jobs=-1, verbose=1)
fine_tune_grid.fit(X_train, y_train)

results_tuned = pd.DataFrame(fine_tune_grid.cv_results_)
best_index = results_tuned['mean_test_score'].idxmax()
best_params = results_tuned.loc[best_index, 'params']

print('best parameters for model:-')
print(best_params)

tuned_model = fine_tune_grid.best_estimator_
y_pred_tuned = tuned_model.predict(X_test)

res_tuned = results(y_test,y_pred_tuned)
res_tuned
```

#### as we can see we are getting both accuracy and f1 score highest after hyperparamter tunning.

```python
cm = confusion_matrix(y_test,y_pred_tuned)

sns.heatmap(cm,annot=True,cmap='Blues',fmt='d',
            xticklabels = np.unique(y_test),
            yticklabels = np.unique(y_pred_tuned))
plt.ylabel('actual label')
plt.xlabel('predicted label')
plt.title('confusion matrix with fine tuned gradient boosting classifier')
plt.show()
```
